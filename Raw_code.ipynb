```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://www.amazon.jobs/en/jobs/3012205/software-development-engineer-ii")

page_content= loader.load().pop().page_content
page_content
```

Creating the templates for resume and job


```python
from langchain_core.prompts import PromptTemplate
```

```python

prompt_extract = PromptTemplate.from_template(
        """
        ### SCRAPED TEXT FROM WEBSITE
        {page_data}
        ### INSTRUCTION
        The scraped test is from the carrer's page of website.
        Your job is to extract the job posting and return them in JSON format contain the 
        following keys : 'company','role','experience','skills',and 'description'.
        Only return the valid JSON.
        ### VALID JSON (NO PREAMBLE):
        """
)

chain_extract = prompt_extract | llm
response = chain_extract.invoke(input={'page_data':page_content})
```

```python
print(response.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser

json_parser = JsonOutputParser()
json_response = json_parser.parse(response.content)
json_response
```

```python
json_response["company"]
```

```python
job_description = json_response    #got job_description in json format
```

```python
print(job_description)
```

```python
prompt_email = PromptTemplate.from_template(
    
    """
    ### RESUME DESCRIPTION
    {resume}
    As a smart skills checker compare the skills,experience,and all required description that allign the {job_description}.
    Analyse and give the view  wiht 'Maching score', 'Required skills', 'Your skills', 'Matching skills' and 'focus/improve skills' concise and on to poin.
    Do not provide preamble 
    ### Response (NO PREAMBEL):
    """

)

chain_email = prompt_email | llm
response = chain_email.invoke({"job_description":str(job_description),"resume": str(resume_content)})
print(response.content)
```

```python
import re

# Example: your LLM output
output_text = response.content

# Use regex to find sections
match_score = re.search(r"\*\*Matching score\*\*: (.+?)\n", output_text).group(1).strip()

required_skills = re.search(
    r"\*\*Required skills\*\*:[\s\S]*?\*\*Your skills", output_text
).group(0)
required_skills = re.findall(r"- (.+)", required_skills)

your_skills = re.search(
    r"\*\*Your skills\*\*:[\s\S]*?\*\*Matching skills", output_text
).group(0)
your_skills = re.findall(r"- (.+)", your_skills)

matching_skills = re.search(
    r"\*\*Matching skills\*\*:[\s\S]*?\*\*Focus/improve skills", output_text
).group(0)
matching_skills = re.findall(r"- (.+)", matching_skills)

focus_skills = re.search(
    r"\*\*Focus/improve skills\*\*:([\s\S]*)", output_text
).group(1)
focus_skills = re.findall(r"- (.+)", focus_skills)

# ✅ Now you have clean lists/strings
print("Matching Score:", match_score)
print('')
print("Required Skills:", required_skills)
print('')
print("Your Skills:", your_skills)
print('')
print("Matching Skills:", matching_skills)
print('')
print("Focus/Improve Skills:", focus_skills)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_coverletter = PromptTemplate.from_template(
    """
    ### JOB DESCRIPTION:
    {job_description}

    ### RESUME:
    {resume}

    Based on the above job description and resume, write a professional and concise cover letter tailored to the job. Focus on alignment of skills, motivation, and impact. Do not include "Dear Sir/Madam" — just start with the letter body.

    ### Cover Letter:
    """
)
```

```python
chain_coverletter = prompt_coverletter | llm

coverletter = chain_coverletter.invoke({
    "job_description": job_description,
    "resume": resume_content
})

print(coverletter.content)
```
